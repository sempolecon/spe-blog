```python
# generar c√≥digo en python para simular una regresi√≥n lineal simple
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from scipy import stats
# Configuraci√≥n de estilo para gr√°ficos
sns.set(style="whitegrid")
# Generar datos sint√©ticos
np.random.seed(42)
n_samples = 100
X = np.random.rand(n_samples) * 10  # Variable independiente
# Relaci√≥n lineal con ruido
y = 3 * X + np.random.randn(n_samples) * 5  # Variable dependiente
# Crear un DataFrame
data = pd.DataFrame({'X': X, 'y': y})
# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(data[['X']], data['y'], test_size=0.2, random_state=42)
# Crear el modelo de regresi√≥n lineal
model = LinearRegression()
model.fit(X_train, y_train)
# Hacer predicciones
y_pred = model.predict(X_test)
# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
print(f'R^2 Score: {r2:.2f}')
# Coeficientes del modelo
print(f'Intercept: {model.intercept_:.2f}')
print(f'Coefficient: {model.coef_[0]:.2f}')
# Visualizar los resultados
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['y'], color='blue', label='Datos reales')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='L√≠nea de regresi√≥n')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresi√≥n Lineal Simple')
plt.legend()
plt.show()
# An√°lisis de residuos
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='purple')
plt.title('Distribuci√≥n de los residuos')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.show()
# Prueba de normalidad de los residuos
k2, p = stats.normaltest(residuals)
alpha = 0.05
print(f'Prueba de normalidad p-value: {p:.3f}')
if p < alpha:
    print("Los residuos no siguen una distribuci√≥n normal (rechazamos H0)")
else:
    print("Los residuos siguen una distribuci√≥n normal (no rechazamos H0)")
```
